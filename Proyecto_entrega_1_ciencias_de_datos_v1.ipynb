## Objetivo general del caso

## Contexto de la base de datos
* Unidad de análisis: Estudiantes que se desvincularon del sistema escolar entre abril y julio de 2022 (DropoutDate).

* Definición de desvinculación: El estudiante ya no aparece en la nómina de ninguna escuela pública.

* Posibles trayectorias:

* **30 de mayo de 2022**: Corte para identificar estudiantes aún no revinculados.

* **28 de junio de 2022**: MINEDUC envía un reporte a las escuelas con los estudiantes desvinculados que aún no se han revinculado (variable ListedInDropoutReport = 1).

* **31 de agosto de 2022**: Se evalúa si el estudiante logró re-enrolarse (EnrolledByAug312022).

* **Fin de 2022**: Se evalúa si el estudiante se graduó (GraduatedIn2022).

```python
# prompt: observar cantidad de estudiantes en el reporte por colegio, observemos la distribución de cantidad de colegios con n cantidad de alumnos reportados, en un histograma solo con mas de 12 estudiantes, con linea horizonal en 1 y 5

import matplotlib.pyplot as plt
#Observar cantidad de estudiantes en el reporte por colegio, observemos la distribución de cantidad de colegios con n cantidad de alumnos reportados, en un histograma solo con mas de 12 estudiantes, con linea horizonal en 1 y 5
students_per_school = df['IDschool'].value_counts()

# Filtrar colegios con más de 12 estudiantes
students_per_school_more_than_12 = students_per_school[students_per_school > 10]

# Crear el histograma filtrado
plt.figure(figsize=(10, 6))
plt.hist(students_per_school_more_than_12, bins=50, edgecolor='black') # Ajustar los bins si es necesario
plt.title('Distribución de Colegios con > 12 Estudiantes Reportados')
plt.xlabel('Cantidad de Estudiantes Reportados')
plt.ylabel('Cantidad de Colegios')
plt.grid(axis='y', alpha=0.75)

# Agregar líneas horizontales
plt.axhline(y=1, color='r', linestyle='--', label='1 colegio')
plt.axhline(y=5, color='g', linestyle='--', label='5 colegios')
plt.axhline(y=15, color='orange', linestyle='--', label='15 colegios')

plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filtrar los colegios con 10 o menos estudiantes reportados
schools_less_than_equal_10_students = students_per_school[students_per_school <= 10].index

# Filtrar el dataframe original para incluir solo a los estudiantes de estos colegios
df_filtered_schools_leq10 = df[df['IDschool'].isin(schools_less_than_equal_10_students)]

# Contar la cantidad total de estudiantes prioritarios en este grupo de colegios
total_prioritary_students_leq10 = df_filtered_schools_leq10['PrioritaryStudentIn2021'].sum()

# Contar la cantidad total de estudiantes en este grupo de colegios
total_students_leq10 = len(df_filtered_schools_leq10)

# Calcular el porcentaje de estudiantes prioritarios en este grupo
percentage_prioritary_students_leq10 = (total_prioritary_students_leq10 / total_students_leq10) * 100 if total_students_leq10 > 0 else 0

print(f"En los colegios con igual o menos de 10 estudiantes reportados:")
print(f"- Cantidad total de estudiantes prioritarios reportados: {total_prioritary_students_leq10}")
print(f"- Cantidad total de estudiantes reportados: {total_students_leq10}")
print(f"- Porcentaje de estudiantes prioritarios reportados: {percentage_prioritary_students_leq10:.2f}%")

# Crear un dataframe para el histograma
data_for_histogram = {
    'Category': ['Prioritarios', 'No Prioritarios'],
    'Cantidad': [total_prioritary_students_leq10, total_students_leq10 - total_prioritary_students_leq10],
    'Porcentaje': [percentage_prioritary_students_leq10, 100 - percentage_prioritary_students_leq10]
}
df_histogram = pd.DataFrame(data_for_histogram)

# Crear el histograma
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(df_histogram['Category'], df_histogram['Cantidad'], color=['skyblue', 'lightcoral'])

ax.set_ylabel('Cantidad de Estudiantes')
ax.set_title('Cantidad y Porcentaje de Estudiantes Prioritarios en Colegios con ≤ 10 Estudiantes Reportados')
ax.set_ylim(0, max(df_histogram['Cantidad']) * 1.2)

# Agregar etiquetas de cantidad y porcentaje a cada barra
for i, bar in enumerate(bars):
    yval = bar.get_height()
    percentage = df_histogram.loc[i, 'Porcentaje']
    plt.text(bar.get_x() + bar.get_width()/2., yval + 10,
             f'{int(yval)}\n({percentage:.1f}%)',
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# prompt: observamos la cantidad de revinculados y en otro de graduados por grado y otro de listados
import pandas as pd
import matplotlib.pyplot as plt

# Agrupaciones por grado escolar
revinculados_por_grado = df.groupby('SchoolGrade2022')['EnrolledByAug312022'].sum().reset_index()
revinculados_por_grado.rename(columns={'EnrolledByAug312022': 'Cantidad Revinculados'}, inplace=True)

graduados_por_grado = df.groupby('SchoolGrade2022')['GraduatedIn2022'].sum().reset_index()
graduados_por_grado.rename(columns={'GraduatedIn2022': 'Cantidad Graduados'}, inplace=True)

listados_por_grado = df.groupby('SchoolGrade2022')['ListedInDropoutReport'].sum().reset_index()
listados_por_grado.rename(columns={'ListedInDropoutReport': 'Cantidad Listados en Reporte'}, inplace=True)

# Merge de los tres indicadores
merged_all = revinculados_por_grado.merge(graduados_por_grado, on='SchoolGrade2022', how='outer')
merged_all = merged_all.merge(listados_por_grado, on='SchoolGrade2022', how='outer')
merged_all = merged_all.fillna(0)
merged_all = merged_all.sort_values('SchoolGrade2022')

# Mostrar tabla resumen
print("Resumen de estudiantes por grado:")
print(merged_all)

# Visualización conjunta
plt.figure(figsize=(14, 7))
x = merged_all['SchoolGrade2022']
width = 0.25

plt.bar(x - width, merged_all['Cantidad Revinculados'], width=width, label='Revinculados', color='skyblue')
plt.bar(x, merged_all['Cantidad Graduados'], width=width, label='Graduados', color='lightgreen')
plt.bar(x + width, merged_all['Cantidad Listados en Reporte'], width=width, label='Listados en Reporte', color='salmon')

plt.xlabel('Grado Escolar')
plt.ylabel('Cantidad de Estudiantes')
plt.title('Revinculados, Graduados y Listados en Reporte por Grado Escolar (2022)')
plt.xticks(x)
plt.legend()
plt.grid(axis='y', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# prompt: metodo del codo para identificar cantidad de clusters ideal para pca

import matplotlib.pyplot as plt
import numpy as np
# Método del codo para PCA
# Determinar el número óptimo de componentes principales

# Calcular la varianza explicada por cada componente
# X_scaled ya está estandarizado desde el paso anterior para PCA+KNN
pca = PCA()
pca.fit(X_scaled)

# Varianza explicada por cada componente
explained_variance_ratio = pca.explained_variance_ratio_

# Varianza explicada acumulada
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Graficar el método del codo
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--', label='Varianza Explicada Individual')
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_explained_variance, marker='o', linestyle='-', color='red', label='Varianza Explicada Acumulada')

plt.title('Método del Codo para PCA')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.grid(True)
plt.legend()
plt.axhline(y=0.95, color='gray', linestyle=':', label='95% Varianza Acumulada') # Línea de referencia para 95%
plt.axhline(y=0.90, color='green', linestyle=':', label='90% Varianza Acumulada') # Línea de referencia para 90%

# Marcar el punto donde se alcanza el 95% o 90%
n_components_95 = np.argmax(cumulative_explained_variance >= 0.95) + 1
n_components_90 = np.argmax(cumulative_explained_variance >= 0.90) + 1

plt.axvline(x=n_components_95, color='gray', linestyle='--', label=f'{n_components_95} comps para 95%')
plt.axvline(x=n_components_90, color='green', linestyle='--', label=f'{n_components_90} comps para 90%')

plt.legend()
plt.show()

print(f"Número de componentes para explicar el 90% de la varianza: {n_components_90}")
print(f"Número de componentes para explicar el 95% de la varianza: {n_components_95}")
```

```python


import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, accuracy_score, f1_score

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# --- DATOS DE ENTRADA ---
# Suponiendo que ya tienes cargado df_2 y la variable target y2
# Y que has definido cols_to_drop_knn
X_ae = df_2.drop(columns=cols_to_drop_knn)
y_ae = y2

# --- ESCALADO DE DATOS ---
scaler_ae = StandardScaler()
X_ae_scaled = scaler_ae.fit_transform(X_ae)

# --- DEFINICIÓN DEL AUTOENCODER ---
input_dim = X_ae_scaled.shape[1]
encoding_dim = 14  # Puedes ajustar esto para más o menos "componentes"

# Capa de entrada
input_layer = Input(shape=(input_dim,))

# Encoder
encoder = Dense(units=64, activation='relu')(input_layer)
encoder = Dense(units=32, activation='relu')(encoder)
encoder = Dense(units=16, activation='relu')(encoder)
encoder_output = Dense(units=encoding_dim, activation='linear')(encoder)  # Latente

# Decoder
decoder = Dense(units=16, activation='relu')(encoder_output)
decoder = Dense(units=32, activation='relu')(decoder)
decoder = Dense(units=64, activation='relu')(decoder)
decoder_output = Dense(units=input_dim, activation='linear')(decoder)  # Reconstrucción

# Modelo completo
autoencoder = Model(inputs=input_layer, outputs=decoder_output)
autoencoder.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

# --- ENTRENAMIENTO DEL AUTOENCODER ---
X_ae_train, X_ae_val = train_test_split(X_ae_scaled, test_size=0.2, random_state=42)
early_stop = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

print("Entrenando autoencoder...")
history = autoencoder.fit(
    X_ae_train, X_ae_train,
    validation_data=(X_ae_val, X_ae_val),
    epochs=200,
    batch_size=32,
    shuffle=True,
    callbacks=[early_stop],
    verbose=0
)
print("Entrenamiento finalizado.")

# --- VISUALIZAR PÉRDIDA ---
plt.figure(figsize=(8, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Pérdida del Autoencoder durante el Entrenamiento')
plt.xlabel('Época')
plt.ylabel('MSE')
plt.legend()
plt.grid(True)
plt.show()

# --- REDUCCIÓN DE DIMENSIONALIDAD (PCA NO LINEAL) ---
encoder_model = Model(inputs=input_layer, outputs=encoder_output)
X_ae_reduced = encoder_model.predict(X_ae_scaled)

print(f"Dimensiones originales: {X_ae_scaled.shape}")
print(f"Dimensiones reducidas (latentes): {X_ae_reduced.shape}")

# --- VISUALIZACIÓN DEL ESPACIO LATENTE (SI encoding_dim == 2 o 3) ---
if encoding_dim == 2:
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=X_ae_reduced[:,0], y=X_ae_reduced[:,1], hue=y_ae, palette='coolwarm')
    plt.title("Representación 2D del espacio latente")
    plt.xlabel("Componente 1")
    plt.ylabel("Componente 2")
    plt.grid(True)
    plt.show()
elif encoding_dim == 3:
    from mpl_toolkits.mplot3d import Axes3D
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(X_ae_reduced[:,0], X_ae_reduced[:,1], X_ae_reduced[:,2], c=y_ae, cmap='coolwarm')
    ax.set_title("Espacio latente 3D")
    plt.show()

# --- CLASIFICACIÓN CON KNN EN EL ESPACIO LATENTE ---
X_ae_train_red, X_ae_test_red, y_ae_train, y_ae_test = train_test_split(
    X_ae_reduced, y_ae, test_size=0.2, random_state=42, stratify=y_ae
)

knn_ae = KNeighborsClassifier(n_neighbors=5)
knn_ae.fit(X_ae_train_red, y_ae_train)

y_pred_ae = knn_ae.predict(X_ae_test_red)
y_proba_ae = knn_ae.predict_proba(X_ae_test_red)[:, 1]

# --- EVALUACIÓN DEL MODELO ---
print("\n=== Reporte de Clasificación ===")
print(classification_report(y_ae_test, y_pred_ae))

print("=== Métricas adicionales ===")
print("Accuracy:", accuracy_score(y_ae_test, y_pred_ae))
print("F1 Score:", f1_score(y_ae_test, y_pred_ae))
print("AUC ROC:", roc_auc_score(y_ae_test, y_proba_ae))

# --- MATRIZ DE CONFUSIÓN ---
cm_ae = confusion_matrix(y_ae_test, y_pred_ae)
sns.heatmap(cm_ae, annot=True, fmt='d', cmap='Purples',
            xticklabels=['No Re-enrolled', 'Re-enrolled'],
            yticklabels=['No Re-enrolled', 'Re-enrolled'])
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión - KNN con Autoencoder')
plt.show()
```

```python
# prompt: hagamos un pca clásico ahora, prcomp o pca() sobnre los datos, primero meotdo del codo para ver lacantidad de componentes y despues aplicamos

import matplotlib.pyplot as plt
import numpy as np
# Método del codo para PCA
# Determinar el número óptimo de componentes principales

# Calcular la varianza explicada por cada componente
# X_scaled ya está estandarizado desde el paso anterior para PCA+KNN
pca = PCA()
pca.fit(X_scaled)

# Varianza explicada por cada componente
explained_variance_ratio = pca.explained_variance_ratio_

# Varianza explicada acumulada
cumulative_explained_variance = np.cumsum(explained_variance_ratio)

# Graficar el método del codo
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o', linestyle='--', label='Varianza Explicada Individual')
plt.plot(range(1, len(explained_variance_ratio) + 1), cumulative_explained_variance, marker='o', linestyle='-', color='red', label='Varianza Explicada Acumulada')

plt.title('Método del Codo para PCA')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada')
plt.xticks(range(1, len(explained_variance_ratio) + 1))
plt.grid(True)
plt.legend()
plt.axhline(y=0.95, color='gray', linestyle=':', label='95% Varianza Acumulada') # Línea de referencia para 95%
plt.axhline(y=0.90, color='green', linestyle=':', label='90% Varianza Acumulada') # Línea de referencia para 90%

# Marcar el punto donde se alcanza el 95% o 90%
n_components_95 = np.argmax(cumulative_explained_variance >= 0.95) + 1
n_components_90 = np.argmax(cumulative_explained_variance >= 0.90) + 1

plt.axvline(x=n_components_95, color='gray', linestyle='--', label=f'{n_components_95} comps para 95%')
plt.axvline(x=n_components_90, color='green', linestyle='--', label=f'{n_components_90} comps para 90%')

plt.legend()
plt.show()

print(f"Número de componentes para explicar el 90% de la varianza: {n_components_90}")
print(f"Número de componentes para explicar el 95% de la varianza: {n_components_95}")

# Aplicar PCA con el número óptimo de componentes (ej: 12 para 90% o 14 para 95%)
# Vamos a elegir 12 componentes como ejemplo
n_components_chosen = n_components_90 # o n_components_95

pca_final = PCA(n_components=n_components_chosen)
X_pca_final = pca_final.fit_transform(X_scaled) # X_scaled viene del paso anterior (PCA+KNN)

print(f"\nDatos originales (escalados) shape: {X_scaled.shape}")
print(f"Datos transformados con PCA shape ({n_components_chosen} componentes): {X_pca_final.shape}")

# Ahora puedes usar X_pca_final para entrenar tu modelo de clasificación o para otros análisis
# Por ejemplo, separar en train/test y entrenar un modelo
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(
    X_pca_final, y_pca, test_size=0.2, random_state=42, stratify=y_pca
)

# Ejemplo con KNN en el espacio PCA reducido
knn_pca_final = KNeighborsClassifier(n_neighbors=5) # Puedes ajustar n_neighbors
knn_pca_final.fit(X_train_pca, y_train_pca)

y_pred_pca_final = knn_pca_final.predict(X_test_pca)
y_proba_pca_final = knn_pca_final.predict_proba(X_test_pca)[:, 1]

print(f"\n=== Reporte de Clasificación (KNN con PCA - {n_components_chosen} componentes) ===")
print(classification_report(y_test_pca, y_pred_pca_final))

print("=== AUC-ROC ===")
print(roc_auc_score(y_test_pca, y_proba_pca_final))

# Matriz de confusión
cm_pca_final = confusion_matrix(y_test_pca, y_pred_pca_final)
sns.heatmap(cm_pca_final, annot=True, fmt='d', cmap='Oranges',
            xticklabels=['No Re-enrolled', 'Re-enrolled'],
            yticklabels=['No Re-enrolled', 'Re-enrolled'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(f'Matriz de Confusión - KNN con PCA ({n_components_chosen} componentes)')
plt.show()

# Puedes repetir este proceso con otros modelos (Regresión Logística, Random Forest, XGBoost)
# utilizando X_pca_final como conjunto de features.
```
